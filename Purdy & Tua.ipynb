<h2>Import libraries

```python
import pandas as pd
#import matplotlib as plt
#import numpy as np
#import seaborn as sns
```

<h2>Import data from website and pass to dataframe

```python
url_passing = 'https://www.fantasypros.com/nfl/advanced-stats-qb.php'
passing = pd.read_html(url_passing)
passing_stats = passing[0]
display(passing_stats.head(10))
```

<h2>Drop 1st level of multi-level index

```python
passing_stats.columns = passing_stats.columns.droplevel(0)
display(passing_stats.head(10))
```

<h2>Subset data to necessary fields only

```python
passing_stats_df = passing_stats[["Player","G","CMP","ATT","PCT","YDS","AIR","AIR/A"]]
display(passing_stats_df.head(10))
```

<h2>Rename a couple of fields here.

```python
passing_stats_df = passing_stats_df.rename(columns={"AIR": "AY", "AIR/A": "AY/ATT"})
display(passing_stats_df.head(10))
```

<h2>Check data types to ensure they are prepared for calculations where necessary

```python
passing_stats_df.dtypes
```

<h2>Divide AY_YDS by Completions (CMP) to create new column, "AY/CMP" (Air Yards Per Completion), filter for players that have at least 150 pass attempts (ATT), display data sorted by AY/CMP in descending order.

```python
passing_stats_df["AY/CMP"] = round(passing_stats_df["AY"] / passing_stats_df["CMP"], 1) 
passing_stats_df = passing_stats_df.query('ATT >= 150')
display(passing_stats_df.sort_values("AY/CMP", ascending=False).head(10))
```

<h2>Display data sorted by AY/ATT in descending order to see how closely the order resembles AY/CMP

```python
display(passing_stats_df.sort_values("AY/ATT", ascending=False).head(10))
```

<h2>Create new fields, ATT_RANK, AY_RANK, AY_RANK, AY/COMP_RANK, AY/ATT_RANK, to rank players in all 4 of those fields.

```python
passing_stats_df['ATT_RANK'] = passing_stats_df['ATT'].rank(method='first', ascending=False).astype(int)
passing_stats_df['AY_RANK'] = passing_stats_df['AY'].rank(method='first', ascending=False).astype(int)
passing_stats_df['AY/CMP_RANK'] = passing_stats_df['AY/CMP'].rank(method='first', ascending=False).astype(int)
passing_stats_df['AY/ATT_RANK'] = passing_stats_df['AY/ATT'].rank(method='first', ascending=False).astype(int)
display(passing_stats_df.sort_values("AY", ascending=False).head(10))
```

<h2>Add air rank fields together to create new field, "AIR_PTS" to assign points for air categories
<h2>Create another field, AIR_PTS_RANK, that ranks players based on AIR_PTS (the lower the points, the higher the ranking).

```python
passing_stats_df['AIR_PTS'] = passing_stats_df['AY_RANK'] + passing_stats_df["AY/CMP_RANK"] + passing_stats_df["AY/ATT_RANK"]
passing_stats_df['AIR_PTS_RANK'] = passing_stats_df['AIR_PTS'].rank(method='first', ascending=True).astype(int)
display(passing_stats_df.sort_values('AIR_PTS_RANK').head(10))
```

<h2>Subset data for only the fields that are air-yards related

```python
passing_stats_final_df = passing_stats_df[['Player','ATT','AY','AY/CMP','AY/ATT','AIR_PTS',
                                           'ATT_RANK','AY_RANK','AY/CMP_RANK','AY/ATT_RANK','AIR_PTS_RANK']]
display(passing_stats_final_df.sort_values('AIR_PTS').head(10))
```
